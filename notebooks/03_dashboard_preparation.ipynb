```python
dashboard_data = predictions[
    [
        "date",
        "family",
        "sales",
        "prediction",
        "error",
        "abs_error",
        "ape",
        "onpromotion",
        "dataset"
    ]
].copy()

dashboard_data.head()
```

```python
dashboard_data.info()
```

```python

dashboard_data = dashboard_data.sort_values(
    by=["family", "date"]
).reset_index(drop=True)

dashboard_data.head()
```

```python
dashboard_data["accuracy"] = (
    100 - dashboard_data["ape"]
)

dashboard_data["accuracy"] = (
    dashboard_data["accuracy"]
    .clip(lower=0, upper=100)
)
```

```python
numeric_cols = [
    "sales",
    "prediction",
    "error",
    "abs_error",
    "ape",
    "accuracy"
]

dashboard_data[numeric_cols] = (
    dashboard_data[numeric_cols]
    .round(2)
)
```
